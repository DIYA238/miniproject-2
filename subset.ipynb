```python
import random as r
from itertools import combinations

a = set([-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4])


print("Subsets of size 5 whose sum is 0:\n")

result1 = []
for subset in combinations(a, 5):
    if sum(subset) == 0:
        result1.append(subset)

for subset in result1:
    sortedelements = sorted(subset)
    print(sortedelements)


result2 = []
for subset in combinations(a, 2):
    if sum(subset) == 0:
        result2.append(subset)

for subset in result2:
    print(subset)


result3 = []
for subset in combinations(a, 3):
    if sum(subset) == 0:
        result3.append(subset)

for subset in result3:
    print(subset)



result4 = []
for subset in combinations(a, 4):
    if sum(subset) == 0:
        result4.append(subset)

for subset in result4:
    print(subset)



result6 = []
for subset in combinations(a, 6):
    if sum(subset) == 0:
        result6.append(subset)

for subset in result6:
    print(subset)


s = len(result1) + len(result2) + len(result3) + len(result4) + len(result6)
print("the number of sets =" ,s)

```
